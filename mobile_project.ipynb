**TARGET TO  PREDICT VALUE **

```python
X=df.drop('price_range',axis=1)
y=df['price_range']
```

**DATA TRAIN_TEST_SPLIT**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.10,random_state = 42)
```

**USING STANDARD SCALER USE**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train =sc.fit_transform(X_train)
X_test =sc.transform(X_test)
```

**CREATE A DEF FUNCTION " EVALUATE" **

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
def evaluate(model,X_train,X_test,y_train,y_test):
    train_pred =model.predict(X_train)
    test_pred =model.predict(X_test)
    print(f"Accuracy for your {model} is: ")
    print(f"R2 score for training {model} is {r2_score(y_train,train_pred)}")
    print(f"R2 score for testing {model} is {r2_score(y_test,test_pred)}")
  #MAE
    print(f"training MAE for {model} is {mean_absolute_error(y_train,train_pred)}")
    print(f"testing MAE for {model} is {mean_absolute_error(y_test,test_pred)}")
  #RMSE
    print(f"training RMSE for {model} is {root_mean_squared_error(y_train,train_pred)}")
    print(f"testing RMSE for {model} is {root_mean_squared_error(y_test,test_pred)}")
```

**FIRST MODEL APPLY LINEAR REGRESSOR **

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
```

```python
le.fit(X_train,y_train)
```

```python
evaluate(le,X_train,X_test,y_train,y_test)
```

**USING GRID SERACHCV**

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
```

```python
ridge_param={
    'alpha':[0.1,1,5,10,50,100]
}
```

```python
rigid_search = GridSearchCV(ridge,ridge_param,cv=5)
```

```python
rigid_search.fit(X_train,y_train)
```

```python
best_ridge =rigid_search.best_estimator_
```

```python
evaluate(best_ridge,X_train,X_test,y_train,y_test)
```

**SECOND MODEL APPLY RANDOMFORESTREGRESSOR**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf_param ={
    'n_estimators':[10,50,100],
    'max_depth':[3,5,7,10],
    'max_features':['sqrt','log2']
}
```

**USING GRID SERACHCV**



```python
rf_search= GridSearchCV(rf,rf_param,cv=5)
```

```python
rf_search.fit(X_train,y_train)
```

```python
best_rf =rf_search.best_estimator_
```

```python
evaluate(best_rf,X_train,X_test,y_train,y_test)
```
